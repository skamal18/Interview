Finding correlation between state and user:


```python
interview=interview.dropna()
interview
```

```python
interview["url"] = interview["url"].astype('category')
interview.dtypes

interview["url_cat"] = interview["url"].cat.codes
interview
```

```python
interview["state"] = interview["state"].astype(float)
interview.info()
```

```python
corr = interview['state'].corr(interview['url_cat'])
corr
```

```python
plt.figure(figsize=(12,12))
plt.title('Relationship between URL and state')
sns.scatterplot(data=interview,x='url',y='state');
```
